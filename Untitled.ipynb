```python
import csv
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter

with open('D:\PythonExercises\Matplotlib\data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    lang_counter = Counter()
    for row in csv_reader:
        lang_counter.update(row['LanguagesWorkedWith'].split(';'))

languages=[]
popularity=[]

languages_list = lang_counter.most_common(15)

for item in languages_list:
    languages.append(item[0])
    popularity.append(item[1])
#this is when the graph requires reverse ordering of data
languages.reverse()
popularity.reverse()

plt.grid()
plt.barh(languages, popularity)

plt.ylabel("Languages")
plt.xlabel("Popularity")
```
